```python
from keras.models import Sequential
from keras.layers import Dense
import numpy as np
```

```python
import pandas as pd
```

# Making random data and labels

```python
data = np.random.random((1000,100))
labels = np.random.randint(2, size = (1000,1))
```

```python
pd.DataFrame(data)
```

```python
pd.DataFrame(labels)
```

# Making a sequential net and adding layers to it

```python
model = Sequential()
model.add(Dense(32,activation='relu', input_dim=100))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(optimizer='rmsprop', loss='binary_crossentropy', metrics=['accuracy'])
```

# Fitting the model

```python
model.fit(data, labels, epochs=10, batch_size=32)
```

```python
model.evaluate(data,labels)
```

# Making Predictions

```python
prediction = model.predict(data)
```

```python
a = pd.DataFrame(prediction,columns=['predicted'])
```

```python
a['actual'] = labels
```

```python
a['predicted_1'] = np.where(a.predicted>=0.5,1,0)
```

```python
from sklearn import metrics 
```

```python
metrics.accuracy_score(list(a.actual),list(a.predicted_1))
```
