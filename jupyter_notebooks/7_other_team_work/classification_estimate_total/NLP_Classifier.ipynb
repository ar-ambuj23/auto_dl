###### Text Cleaning

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction import DictVectorizer
from scipy.sparse import hstack
```

```python
# For beginning, transform train['FullDescription'] to lowercase using text.lower()
filtered_df['LossDescription'].str.lower()

# Then replace everything except the letters and numbers in the spaces.
# it will facilitate the further division of the text into words.
filtered_df['LossDescription'].replace('[^a-zA-Z0-9]', ' ', regex = True)
```

```python
def remove_num(text):
    text = ''.join([i for i in text if not i.isdigit()])
    return text
filtered_df['LossDescription'] = np.vectorize(remove_num)(filtered_df['LossDescription'])
```

```python
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
def remove_stopwords(text):
    word_list = text.split()
    word_list = [word for word in word_list if word not in stop_words]
    text = ' '.join(word_list)
    return text
filtered_df['LossDescription'] = np.vectorize(remove_stopwords)(filtered_df['LossDescription'])
```

```python
from string import punctuation

def remove_punct(text):
    text = ' '.join(word.strip(punctuation) for word in text.split() if word.strip(punctuation))
    return text
filtered_df['LossDescription'] = np.vectorize(remove_punct)(filtered_df['LossDescription'])

def remove_u(text):
    text = text.replace('_','')
    return text
filtered_df['LossDescription'] = np.vectorize(remove_u)(filtered_df['LossDescription'])
```

```python
import itertools
def remove_iter(text):
    text = ''.join(''.join(s)[:2] for _, s in itertools.groupby(text))
    return text
filtered_df['LossDescription'] = np.vectorize(remove_iter)(filtered_df['LossDescription'])
```

```python
def remove_extra_space(text):
    word_list = text.split()
    text = ' '.join(word_list)
    return text
filtered_df['LossDescription'] = np.vectorize(remove_extra_space)(filtered_df['LossDescription'])
```

```python
import re
def expand_contractions(text, contraction_mapping=CONTRACTION_MAP):
    
    contractions_pattern = re.compile('({})'.format('|'.join(contraction_mapping.keys())), 
                                      flags=re.IGNORECASE|re.DOTALL)
    def expand_match(contraction):
        match = contraction.group(0)
        first_char = match[0]
        expanded_contraction = contraction_mapping.get(match)\
                                if contraction_mapping.get(match)\
                                else contraction_mapping.get(match.lower())                       
        expanded_contraction = first_char+expanded_contraction[1:]
        return expanded_contraction
        
    expanded_text = contractions_pattern.sub(expand_match, text)
    expanded_text = re.sub("'", "", expanded_text)
    return expanded_text
filtered_df['LossDescription'] = np.vectorize(expand_contractions)(filtered_df['LossDescription'])
```

```python
from nltk.stem import WordNetLemmatizer

w_tokenizer = nltk.tokenize.WhitespaceTokenizer()
lemmatizer = nltk.stem.WordNetLemmatizer()
```

```python
def lemmatize_text(text):
    word_list = [lemmatizer.lemmatize(w) for w in w_tokenizer.tokenize(text)]
    text= ' '.join(word_list)
    return text
filtered_df['LossDescription'] = np.vectorize(lemmatize_text)(filtered_df['LossDescription'])
```

```python
from nltk.stem.porter import PorterStemmer
stemmer = PorterStemmer()

def stem_text(text):
    processed_tokens = []
    tokens = nltk.word_tokenize(text)
    porter = nltk.PorterStemmer()
    for t in tokens:
        t = porter.stem(t)
        processed_tokens.append(t)
    text = " ".join(processed_tokens)
    return text
filtered_df['LossDescription'] = np.vectorize(stem_text)(filtered_df['LossDescription'])
```
