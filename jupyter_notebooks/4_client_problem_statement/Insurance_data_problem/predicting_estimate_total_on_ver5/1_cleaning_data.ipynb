# STAGE 3 - Treatment of CommOrRess Columns

Convert into only 3 categories: Comm, Res and other

```python
# Reference
# http://jonathansoma.com/lede/algorithms-2017/classes/fuzziness-matplotlib/fuzzing-matching-in-pandas-with-fuzzywuzzy/
```

```python
stage3_df = stage2_df.copy()
```

```python
stage3_df.CommOrRes.value_counts()
```

```python
# fuzz is used to compare TWO strings
from fuzzywuzzy import fuzz

# process is used to compare a string to MULTIPLE other strings
from fuzzywuzzy import process
```

```python
choices = stage3_df['CommOrRes'].unique()
choices[:15]
```

```python
len(choices)
```

#### Residential

```python
residential_possibilities = process.extract("Residential", choices, limit=500, scorer=fuzz.token_set_ratio)
```

```python
residential_possibilities
```

```python
residential_list = [residential_possible for residential_possible in residential_possibilities if residential_possible[1] > 60]
residential_list
```

```python
residential_list_names= [i[0] for i in residential_list]
```

```python
stage3_df_residential = stage3_df[stage3_df.isin({'CommOrRes': residential_list_names})['CommOrRes']]
```

```python
l1 =list(stage3_df_residential.index)
```

#### Commercial

```python
commercial_possibilities = process.extract("Commercial", choices, limit=500, scorer=fuzz.token_set_ratio)
```

```python
commercial_possibilities
```

```python
commercial_list = [commercial_possible for commercial_possible in commercial_possibilities if commercial_possible[1] > 60]
commercial_list
```

```python
commercial_list_names= [i[0] for i in commercial_list]
```

```python
stage3_df_commercial = stage3_df[stage3_df.isin({'CommOrRes': commercial_list_names})['CommOrRes']]
```

```python
l2= list(stage3_df_commercial.index)
```

#### Others

```python
others_list = list(set(l1).intersection(set(l2)))
```

```python
l1.remove(others_list[0])
```

```python
l1.remove(others_list[1])
l1.remove(others_list[2])
```

```python
l2.remove(others_list[0])
l2.remove(others_list[1])
l2.remove(others_list[2])
```

```python
res_df = stage3_df[stage3_df.index.isin(l1)].copy()
```

```python
res_df['CommOrRes']= 'Residential'
```

```python
com_df = stage3_df[stage3_df.index.isin(l2)].copy()
```

```python
com_df['CommOrRes'] = 'Commercial'
```

```python
other_df = stage3_df[~stage3_df.index.isin(l1)].copy()
```

```python
other_df = other_df[~other_df.index.isin(l2)]
```

```python
other_df['CommOrRes'] = 'Other'
```

```python
stage3_df.shape
```

```python
stage3_df1 = pd.concat([res_df,com_df,other_df])
```

```python
stage3_df1.shape
```

```python
stage3_df1.CommOrRes.value_counts()
```
