# Tables

```python
orders = pd.read_csv('data/Orders.csv',index_col=0)
order_regions = pd.read_csv('data/OrderRegions.csv',index_col=0)
order_details = pd.read_csv('data/Order Details.csv',index_col=0)
products = pd.read_csv('data/Products.csv',index_col=0)
categories = pd.read_csv('data/Categories.csv',index_col=0)
shippers = pd.read_csv('data/Shippers.csv',index_col=0)
suppliers = pd.read_csv('data/Suppliers.csv',index_col=0)
customers = pd.read_csv('data/Customers.csv',index_col=0)
```

# Data Preparation

```python
orders.drop(columns=['ShipRegion','ShipName','ShipAddress'],inplace=True)
```

```python
products = products[['ProductID','SupplierID','CategoryID']]
```

```python
order_details1 = pd.merge(order_details[['OrderID','ProductID']],products,left_on='ProductID',right_on='ProductID')
```

```python
order_details1
```

```python
order_details2 = pd.get_dummies(order_details1, columns=['ProductID','CategoryID','SupplierID']).groupby(['OrderID'], as_index=False).sum()
```

```python
final_df = pd.merge(orders,order_details2,left_on='OrderID',right_on='OrderID')
```

```python
final_df = final_df.rename(columns={'EmployeeID':'EmployeeID:string','ShipVia':'ShipVia:string'})
```

```python
final_df.to_csv('northwind_data.csv',index=False)
```
