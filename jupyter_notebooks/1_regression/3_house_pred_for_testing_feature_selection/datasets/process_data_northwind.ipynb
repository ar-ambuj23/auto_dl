```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('northwind_data.csv',index_col=0)
```

```python
def splitter(days):

    try:
        days = int(str(days).split(' ')[0])
    except:
        days = np.nan

    return days
```

```python
df['req_days'] = pd.to_datetime(df.RequiredDate) - pd.to_datetime(df.OrderDate)
df['req_days'] = df.req_days.apply(splitter)
```

```python
df.drop(columns=['OrderDate','RequiredDate','ShippedDate','ShipCity','ShipPostalCode'],inplace=True)
```

```python
df.to_csv('northwind_data_1.csv',index=False)
```
